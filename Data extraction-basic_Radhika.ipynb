```python
def recursively_translate(node):
    if isinstance(node, str):
        if node.strip() != '':
            try:
                # Translate the string and return the translated text
                return translator.translate(node, dest='en').text
            except Exception as e:
                print(f"Translation failed: {e}")
                return node  # Return the original node if translation fails
    else:
        for content in node.contents:
            if isinstance(content, str) and content.strip() != '':
                try:
                    # Translate the string and replace it in the node
                    translated_text = translator.translate(content, dest='en').text
                    content.replace_with(translated_text)
                except Exception as e:
                    print(f"Translation failed: {e}")
            elif content is not None:
                recursively_translate(content)
# Define a function to translate the content of a webpage
def translate_webpage(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Recursively translate the content of the soup object
        recursively_translate(soup)

        # Return the translated soup object
        return soup.prettify()
    else:
        print(f"Error: Unable to fetch the webpage. Status code: {response.status_code}")
        return None
```

```python
from bs4 import BeautifulSoup
import requests
from googletrans import Translator

def translate_webpage(url, target_language="en"):
    try:
        response = requests.get(url)
        response.raise_for_status()

        soup = BeautifulSoup(response.content, "html.parser")
        translator = Translator()
        translatable_tags = ['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'li', 'a', 'span']

        for tag in translatable_tags:
            for element in soup.find_all(tag):
                text = element.get_text(strip=True)
                if text:
                    translation = translator.translate(text, dest=target_language).text
                    element.string = translation

        return str(soup)

    except requests.exceptions.RequestException as e:
        print(f"Error fetching webpage: {e}")
        return None

# Example usage
url = 'https://www.upl-ltd.com/br/defensivos-agricolas'
translated_html = translate_webpage(url, target_language="en")

if translated_html:
    print(translated_html)
else:
    print("Error translating webpage.")
```
