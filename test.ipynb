```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

url = 'https://www.upl-ltd.com/es/Productos'
driver.get(url)

# Find all accordion headings and make sure they are clickable
accordion_headings = WebDriverWait(driver, 30).until(
    EC.presence_of_all_elements_located(
        (By.CSS_SELECTOR, 'div.panel-heading.products-heading[data-toggle="collapse"]'))
)

all_data = []

for i, accordion_heading in enumerate(accordion_headings):
    # Use js to click the accordion heading to expand it
    driver.execute_script("arguments[0].click();", accordion_heading)
    
    # Fetch the category name
    accordion_text = accordion_heading.find_element(By.CSS_SELECTOR, 'h2').text.strip()
    print(f"Fetching data for category: {accordion_text}")
    
    # Wait for the products container to be visible
    try:
        products_container = WebDriverWait(driver, 30).until(
            EC.visibility_of_element_located(
                (By.CSS_SELECTOR, 'div.panel-collapse.collapse.in'))
        )
    except Exception as e:
        print(f"Timeout while waiting for products container for category: {accordion_text}. Error: {e}")
        continue
    
    # Make sure all hidden product rows are visible
    driver.execute_script("""
        var hiddenElements = document.querySelectorAll('.products-row.hidden');
        hiddenElements.forEach(function(elem) {
            elem.classList.remove('hidden');
            elem.style.display = 'block';
        });
    """)
    
    # Find all product elements
    products = products_container.find_elements(By.CSS_SELECTOR, 'div.product')
    p_names = []
    p_ings = []
    p_links = []
    
    for product in products:
        try:
            name = product.find_element(By.CSS_SELECTOR, 'div.product-name h3')
            product_name = name.text.strip()
            if product_name:
                p_names.append(product_name)
            else:
                print("Product name is empty or blank, skipping.")
                continue
        except Exception as e:
            print(f"Error extracting product name: {e}")
            continue

        try:
            # Ensure all ingredients are visible
            driver.execute_script("arguments[0].scrollIntoView();", product)
            ingredient_elements = product.find_elements(By.CSS_SELECTOR, 'div.col-sm-8.col-xs-12 > div > p.product-info > span')
            ingredients = [ing.text.strip() for ing in ingredient_elements]
            if ingredients:
                p_ings.append(ingredients)
            else:
                p_ings.append(["N/A"])
                print("No ingredients found, adding 'N/A'.")
        except Exception as e:
            print(f'Error fetching active ingredients list: {e}')
            p_ings.append(["N/A"])
            continue

        try:
            link_element = product.find_element(By.CSS_SELECTOR, 'a.view-btn')
            product_link = link_element.get_attribute('href')
            p_links.append(product_link)
        except Exception as e:
            print(f"Error extracting product link: {e}")
            p_links.append("N/A")
            continue

    temp_df = pd.DataFrame({'Category': [accordion_text] * len(p_names), 'Product Name': p_names, 'Active ingredients': p_ings, 'Product link': p_links})
    display(temp_df)
    all_data.append(temp_df)

driver.quit()
country_df = pd.concat(all_data)

# Display the final DataFrame
display(country_df)
```
