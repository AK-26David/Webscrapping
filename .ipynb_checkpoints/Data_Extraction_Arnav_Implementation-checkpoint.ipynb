```python
import requests
from bs4 import BeautifulSoup

# URL of the website
URL = "https://www.upl-ltd.com/"

try:
    # Fetch the content of the website
    r = requests.get(URL)
    r.raise_for_status()  # Raise an HTTPError for bad responses
except requests.exceptions.RequestException as e:
    print(f"Error fetching the URL: {e}")
    exit()

# Parse the website content
soup = BeautifulSoup(r.text, 'html.parser')  # Changed 'lxml' to 'html.parser'

# Dictionary to store country URLs
country_url_dict = {}

# Loop through each alphabet character to find corresponding country links
for start in 'abcdefghijklmnopqrstuvwxyz':
    class_name = f"country_wrapper d-flex align-items-center text-decoration-none py-3 all {start}-search"
    try:
        elements = soup.find_all('a', class_=class_name)
        if not elements:
            print(f"No elements found for class: {class_name}")
        for i in elements:
            try:
                # Extract the country name
                country = i.find('h4', class_="country_name pl-2 mb-0").text.strip()
                # Extract the URL
                url = i['href']
                # Store the country name and URL in the dictionary
                country_url_dict[country] = url
            except AttributeError as e:
                print(f"Error extracting data from element: {e}")
    except Exception as e:
        print(f"Error finding elements for class '{class_name}': {e}")

# Print the dictionary with country URLs
for country, url in country_url_dict.items():
    print(f"{country}: {url}")
```

```python
# Creating country homepage url dataframe
df=pd.DataFrame({'Country':list(country_url_dict.keys()),'Url':list(country_url_dict.values())})
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import googletrans
from googletrans import Translator
translator = Translator()  # Ensure this is correctly initialized

for country in df['Country'].unique():
    print(f"Processing country: {country}")
    
    url = country_url_dict.get(country)
    if not url:
        print(f"No URL found for country: {country}")
        continue
    
    print(f"URL: {url}")

    try:
        # Fetch the page content
        page = requests.get(url)
        page.raise_for_status()  # Raise an error for bad status codes
        
        # Translate the page content
        translated = translator.translate(page.text)
        soup = BeautifulSoup(translated.text, 'lxml')
        
        # Get the language code for the country
        lan_code = df.loc[df['Country'] == country, 'lan_code1'].iloc[0]
        print(f"Language code: {lan_code}")
        
        # Translate the word "Products"
        tv = translator.translate('Products', dest=lan_code)
        target_text = tv.text.split(' ')[-1].title() if len(tv.text.split(' ')) > 1 else tv.text
        print(f"Target text: {target_text}")
        
        # Find all links containing the translated word
        matches = soup.find_all('a', string=lambda text: text and target_text in text.strip(), href=True)
        print(f"Found {len(matches)} matches for target text '{target_text}'")

        # Loop through matches to find the product page URL
        product_page_url = None
        for match in matches:
            parent_tag = match.parent
            product_page_url = parent_tag.find('a')['href']
            if product_page_url:
                df.loc[df['Country'] == country, 'Product_page_url'] = product_page_url
                print(f"Product page URL: {product_page_url}")
                break  # Assuming you want only the first match
        else:
            print("No matching product page links found.")
        
        print('---------------')
    except requests.RequestException as e:
        print(f"Request error for {country}: {e}")
    except Exception as e:
        print(f"Error processing {country}: {e}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
import pandas as pd
from time import sleep
from bs4 import BeautifulSoup

# Set up Selenium with headless Chrome options
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')

# URL to scrape
url = 'https://www.upl-ltd.com/es/Productos'

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)

try:
    # Open the URL
    driver.get(url)

    # Wait for the outer table to be fully loaded
    WebDriverWait(driver, 30).until(
        EC.presence_of_all_elements_located((By.XPATH, '//div[@id="product-accordion"]//div[contains(@class, "panel-heading")]'))
    )

    # Extract category names and counts
    categories = driver.find_elements(By.XPATH, '//div[@id="product-accordion"]//div[contains(@class, "panel-heading")]')

    # Debug: Print categories
    for cat in categories:
        print("Category Raw Text:", repr(cat.text))

    # Create a dictionary to store the category count
    category_count = {}
    for cat in categories:
        text = cat.text.strip()
        lines = text.split('\n')
        if len(lines) >= 2:
            category_name = lines[0].strip()
            count_text = lines[-1].strip()
            count = int(count_text.strip('()'))
            category_count[category_name] = count

    # Create a Pandas DataFrame from the category_count dictionary
    df_categories = pd.DataFrame(list(category_count.items()), columns=['Category', 'Count'])

    # Print the DataFrame
    print(df_categories)

    # Wait for a few seconds to ensure the page is fully loaded
    sleep(5)

    # Get the page source
    ps = driver.page_source

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(ps, 'html.parser')

    # Extract the product category
    category_element = soup.find('h2')
    if category_element:
        category = category_element.get_text(strip=True).split()[0]
        print(f"Category: {category}")
    else:
        print("No category found")

    # Find all product elements
    products = soup.find_all('div', class_='product')
    print(f"Number of products found: {len(products)}")

    # Prepare lists to store product details
    product_names = []
    active_ingredients = []

    # Iterate through each product and extract details
    for product in products:
        # Extract product name
        product_name_element = product.find('h3')
        if product_name_element:
            product_name = product_name_element.get_text(strip=True)
            product_names.append(product_name)
            print(f"Product Name: {product_name}")
        else:
            product_names.append("No product name found")

        # Extract active ingredients
        product_info_element = product.find('p', class_='product-info')
        if product_info_element:
            ingredients = product_info_element.find_all('span', class_='product-ingrd')
            active_ingredient_list = [ingredient.get_text(strip=True).split('/')[0].strip() for ingredient in ingredients]
            active_ingredients.append(', '.join(active_ingredient_list))
            print(f"Active Ingredients: {', '.join(active_ingredient_list)}")
        else:
            active_ingredients.append("No active ingredients found")

    # Create a DataFrame to store the data
    df_products = pd.DataFrame({
        'Category': [category] * len(product_names),
        'Product Name': product_names,
        'Active Ingredients': active_ingredients
    })

    # Print the DataFrame
    print(df_products)

finally:
    # Close the WebDriver
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, ElementClickInterceptedException
from selenium.webdriver.chrome.options import Options
import pandas as pd
from time import sleep
from bs4 import BeautifulSoup

# Set up Selenium with headless Chrome options
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')

# URL to scrape
url = 'https://www.upl-ltd.com/es/Productos'

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)

try:
    # Open the URL
    driver.get(url)

    # Wait for the outer table to be fully loaded
    WebDriverWait(driver, 30).until(
        EC.presence_of_all_elements_located((By.XPATH, '//div[@id="product-accordion"]//div[contains(@class, "panel-heading")]'))
    )

    # Extract category names and counts
    categories = driver.find_elements(By.XPATH, '//div[@id="product-accordion"]//div[contains(@class, "panel-heading")]')

    # Debug: Print categories
    for cat in categories:
        print("Category Raw Text:", repr(cat.text))

    # Create a dictionary to store the category count
    category_count = {}
    for cat in categories:
        text = cat.text.strip()
        lines = text.split('\n')
        if len(lines) >= 2:
            category_name = lines[0].strip()
            count_text = lines[-1].strip()
            count = int(count_text.strip('()'))
            category_count[category_name] = count

    # Create a Pandas DataFrame from the category_count dictionary
    df_categories = pd.DataFrame(list(category_count.items()), columns=['Category', 'Count'])

    # Print the DataFrame
    print(df_categories)

    # Prepare lists to store product details
    category_names = []
    product_names = []
    active_ingredients = []
    product_urls = []

    # Iterate through each category and extract product details
    for category in categories:
        category_name = category.text.split('\n')[0].strip()

        # Ensure the element is clickable by scrolling into view
        driver.execute_script("arguments[0].scrollIntoView(true);", category)

        # Try to click the category, retrying if necessary
        for attempt in range(3):
            try:
                # Check if the element is visible and clickable
                WebDriverWait(driver, 10).until(EC.element_to_be_clickable(category))
                category.click()
                sleep(2)  # Wait for the products to load
                break
            except (ElementClickInterceptedException, TimeoutException) as e:
                print(f"Attempt {attempt+1}: Failed to click on category: {category_name}. Error: {e}")
                sleep(2)
                continue
        else:
            print(f"Failed to click on category: {category_name} after 3 attempts.")
            continue

        # Get the updated page source
        ps = driver.page_source
        soup = BeautifulSoup(ps, 'html.parser')

        # Find all product elements within the expanded category
        products_section = soup.find('div', class_='panel-collapse in')
        if products_section:
            products = products_section.find_all('div', class_='product')
        else:
            print(f"No products found for category: {category_name}")
            continue

        # Iterate through each product and extract details
        for product in products:
            # Extract product name
            product_name_element = product.find('h3')
            if product_name_element:
                product_name = product_name_element.get_text(strip=True)
                product_names.append(product_name)
            else:
                product_names.append("No product name found")

            # Extract active ingredients
            product_info_element = product.find('p', class_='product-info')
            if product_info_element:
                ingredients = product_info_element.find_all('span', class_='product-ingrd')
                active_ingredient_list = [ingredient.get_text(strip=True).split('/')[0].strip() for ingredient in ingredients]
                active_ingredients.append(', '.join(active_ingredient_list))
            else:
                active_ingredients.append("No active ingredients found")

            # Extract product URL
            product_link_element = product.find('a', href=True)
            if product_link_element:
                product_url = 'https://www.upl-ltd.com' + product_link_element['href']
                product_urls.append(product_url)
            else:
                product_urls.append("No product URL found")

            # Append the current category name
            category_names.append(category_name)

finally:
    # Close the WebDriver
    driver.quit()

# Create a DataFrame to store the data
df_products = pd.DataFrame({
    'Category': category_names,
    'Product Name': product_names,
    'Active Ingredients': active_ingredients,
    'Product URL': product_urls
})

# Print the DataFrame
print(df_products)
```
