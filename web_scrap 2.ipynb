```python
# Product categories
act_subs = []
for i in soup.find_all('h2'):
       act_subs.append(i.text.strip())
```

```python
# Category and count of products
category_counts=[i.replace('\n','').replace(')','').replace(' ','').split('(') for i in act_subs if '(' and')'in i]
category_counts=list(map(lambda x: (x[0],int(x[-1])),category_counts))

```

```python
category_list_=[]
for i in category_counts:
#     print(i[0],i[-1])
    res= [i[0]]*i[1]
    category_list_.extend(res)
```

```python
# Getting list of products
product = []
for i in soup.find_all('h3'):
    product.append(i.text.strip())
print(product)
```

```python
# # Getting active ingredients of the products
ingredients=[]
for i in soup.find_all('p',class_='product-info'):
    ingredients.append(i.text.strip())
ingredients
```

```python
# Product url links
link = soup.find_all('a',class_="btn upl-btn view-btn")
product_link = []
for i in link:
    product_link.append(i['href'])
print(product_link)
```

```python
product_category_df=pd.DataFrame({'Category':category_list_,'Products':product,'Product_link':product_link})
product_category_df['Product_link']=url.split('.com')[0]+'.com'+product_category_df['Product_link']
product_category_df['Country']=country
# product_category_df.to_excel(f'{country}.xlsx', index=False)
display(product_category_df)
```
